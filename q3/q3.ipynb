```python
import csv
import matplotlib.pyplot as plt

#csv 파일 읽어오기
yearIdx = 2008
data = []
xLabel = []
for i in range(0, 16, 2):
    xLabel.append(str(yearIdx + i))
    yearStr = str(yearIdx + i) + '.csv'
    f = open(yearStr, 'r', encoding='ANSI')
    data.append(csv.reader(f))

results = []

for i in range(len(data)):
    header = next(data[i])
    temp = 0
    row = next(data[i])
    sum = 0
    for k in range(0,12):
        man = int(row[1 + (k * 4) + 2])
        woman = int(row[1 + (k * 4) + 3])
        temp = man - woman
        sum += temp
    average = round(sum/12,0)
    results.append(average)
    
plt.rc('font', family = 'Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.plot(xLabel,results)
    
plt.title('년도별 1월~12월의 남성과 여성 인구수의 차이 평균 ')
plt.xlabel("year")
plt.ylabel("difference (남성 수 - 여성 수)")
plt.show()
```
