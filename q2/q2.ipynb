```python
import random
import csv
import matplotlib.pyplot as plt

#   
trials = [100, 1000, 10000, 100000]

#     
results = []
for i in range(len(trials)):
    results.append([])

# trials   
for i in range(len(trials)):
    for j in range(trials[i]):
        result = random.randint(1, 6)
        results[i].append(result)

#   csv  
with open('q2.csv', 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['trial', 'result'])
    for i in range(len(trials)):
        for j in range(trials[i]):
            writer.writerow([trials[i], results[i][j]])

#  
for i in range(len(trials)):
    plt.figure()
    plt.hist(results[i], bins=6,edgecolor='black')
    plt.title(f'Dice Simulation ({trials[i]} trials)')
    plt.xlabel('Number')
    plt.ylabel('Frequency')
    plt.show()
    
```
