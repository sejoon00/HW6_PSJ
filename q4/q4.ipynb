```python
import csv
import matplotlib.pyplot as plt

#csv 파일 읽어오기

f = open('q4.csv', 'r', encoding='ANSI')
data = csv.reader(f)
header = next(data)
resultIn = {}
resultOut = {}
resultTotal = {}

#승하차 인원 구하기
for row in data:
    name = row[2]
    if name in resultIn:
        resultIn[name] += int(row[9]) + int(row[11])
        resultOut[name] += int(row[10]) + int(row[12])
        resultTotal[name] += int(row[9]) + int(row[11]) + int(row[10]) + int(row[12])
        
    else:
        resultIn[name] = int(row[9]) + int(row[11])
        resultOut[name] = int(row[10]) + int(row[12])
        resultTotal[name] = int(row[9]) + int(row[11]) + int(row[10]) + int(row[12])
    
#정렬
resultIn = sorted(resultIn.items(), key=lambda x: x[1], reverse=True)[:30]
resultOut = sorted(resultOut.items(), key=lambda x: x[1], reverse=True)[:30]
resultTotal = sorted(resultTotal.items(), key=lambda x: x[1], reverse=True)[:30]

#막대 그래프 출력 함수
def printGraph(title, dict):
    plt.rc('font', family='Malgun Gothic')
    plt.figure(figsize=(12, 8))
    plt.title('출근 시간대 최대 '+title+'역 TOP 30')
    plt.xlabel(title+'역')
    plt.ylabel('승객수')
    plt.xticks(rotation=90)
    x = [i[0] for i in dict]
    y = [i[1] for i in dict]
    plt.bar(x, y)
    plt.tight_layout()
    plt.show()


printGraph('승차',resultIn)
printGraph('하차',resultOut)
printGraph('승하차',resultTotal)
```
