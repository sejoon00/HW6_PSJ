```python
import csv
import matplotlib.pyplot as plt

#csv  

f = open('q4.csv', 'r', encoding='ANSI')
data = csv.reader(f)
header = next(data)
resultIn = {}
resultOut = {}
resultTotal = {}

#  
for row in data:
    name = row[2]
    if name in resultIn:
        resultIn[name] += int(row[9]) + int(row[11])
        resultOut[name] += int(row[10]) + int(row[12])
        resultTotal[name] += int(row[9]) + int(row[11]) + int(row[10]) + int(row[12])
        
    else:
        resultIn[name] = int(row[9]) + int(row[11])
        resultOut[name] = int(row[10]) + int(row[12])
        resultTotal[name] = int(row[9]) + int(row[11]) + int(row[10]) + int(row[12])
    
#
resultIn = sorted(resultIn.items(), key=lambda x: x[1], reverse=True)[:30]
resultOut = sorted(resultOut.items(), key=lambda x: x[1], reverse=True)[:30]
resultTotal = sorted(resultTotal.items(), key=lambda x: x[1], reverse=True)[:30]

#   
def printGraph(title, dict):
    plt.rc('font', family='Malgun Gothic')
    plt.figure(figsize=(12, 8))
    plt.title('   '+title+' TOP 30')
    plt.xlabel(title+'')
    plt.ylabel('')
    plt.xticks(rotation=90)
    x = [i[0] for i in dict]
    y = [i[1] for i in dict]
    plt.bar(x, y)
    plt.tight_layout()
    plt.show()


printGraph('',resultIn)
printGraph('',resultOut)
printGraph('',resultTotal)
```
